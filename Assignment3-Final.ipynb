Because the train data set is unbalanced, we undersample the data set and make it balanced.

```python
#Undersample
np.random.seed(12345)
nclaim_indices = train[train.Claim_Amount == 0].index
sample_size = sum(train.Claim_Amount  == 1)
random_indices = np.random.choice(nclaim_indices, sample_size, replace=False)
nclaim_sample = train.loc[random_indices]
claim_sample=train[train.Claim_Amount == 1]
train=pd.concat([nclaim_sample,claim_sample])
```

We planned to create numeric features for the “Blind_Make”, “Blind_Model”, and “Blind_Submodel” by utilizing the method of encoding. We defined a function called “woe” that basically did the following steps:


```python
from math import log
from numpy.random import uniform

def woe(df,columns,target="Claim_Amount"):
    events = len(df[df[target]==1])
    nevents=len(df[df[target]==0])
    for col_name in columns:
        new_col_name = col_name + '_encode'
        row_val_dict = {}
        for index, row in pd.crosstab(df[col_name],df[target]).iterrows():
            row_val_dict[index] = (log(((row[1]+0.5)/events)/((row[0]+0.5)/nevents)))
        df[new_col_name] = df[col_name].apply(lambda i: row_val_dict[i]+ uniform(low=-0.07, high=0.07))
        del df[col_name]
```

```python
import h2o 
from h2o.estimators.gbm import H2OGradientBoostingEstimator 
from h2o.estimators.random_forest import H2ORandomForestEstimator
from h2o.grid.grid_search import H2OGridSearch
```

```python
# start and connect to h2o server
h2o.init(min_mem_size='3G')
```

```python
# location of "dirty" file
# decision trees handle dirty data elegantly
trainpath = 'C:/DataMining/Assignment3/train.csv'
testpath = 'C:/DataMining/Assignment3/test.csv'
validationpath = 'C:/DataMining/Assignment3/validation.csv'
```

```python
col_types = {'Claim_Amount': 'enum'}
```

```python
train= h2o.import_file(path=trainpath, col_types=col_types)
```

```python
train.describe()
```

```python
validation= h2o.import_file(path=validationpath, col_types=col_types)
```

```python
validation.describe()
```

```python
test= h2o.import_file(path=testpath, col_types=col_types)
```

```python
test.describe()
```

```python
y = "Claim_Amount"
X = [name for name in train.columns if name not in ['C1','Row_ID', 'Household_ID','Unnamed: 0', y]]
print(y)
print(X)
```

```python
# set target to factor - for binary classification
train[y] = train[y].asfactor()
validation[y] = validation[y].asfactor()
test[y] = test[y].asfactor()
```

```python
# GBM with random hyperparameter search
# train many different GBM models with random hyperparameters
# and select best model based on validation error

# define random grid search parameters
hyper_parameters = {'ntrees':list(range(0, 400, 2)),
                    'max_depth':list(range(1, 24, 1)),
                    'nbins':list(range(1, 60, 1)),
                    'nbins_cats':list(range(1, 1024, 2)),
                    'sample_rate':[s/float(10) for s in range(1, 11)],
                    'col_sample_rate':[s/float(10) for s in range(1, 11)]}

# define search strategy
search_criteria = {'strategy':'RandomDiscrete',
                   'max_models':20,
                   'max_runtime_secs':600}

# initialize grid search
gsearch = H2OGridSearch(H2OGradientBoostingEstimator,
                        hyper_params=hyper_parameters,
                        search_criteria=search_criteria)

# execute training w/ grid search
gsearch.train(x=X,
              y=y,
              training_frame=train,
              validation_frame=validation)

# view detailed results at http://localhost:54321/flow/index.html
```

```python
# show grid search results
gsearch.show()

# select best model
gbm_model = gsearch.get_grid()[0]

# print model information
gbm_model
```

```python
# measure gbm AUC
print(gbm_model.auc(train=True))
print(gbm_model.auc(valid=True))
print(gbm_model.model_performance(test_data=test).auc())
```

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
```

```python
# NN with random hyperparameter search
# train many different NN models with random hyperparameters
# and select best model based on validation error

# define random grid search parameters
hyper_parameters = {'hidden':[s for s in range(20, 100, 2)],
                    'l1':[s/1e5 for s in range(0, 1000, 2)],
                    'l2':[s/1e5 for s in range(0, 1000, 2)],
                    'input_dropout_ratio':[s/1e3 for s in range(0, 200, 2)]}

# define search strategy
search_criteria = {'strategy':'RandomDiscrete',
                   'max_models':20,
                   'max_runtime_secs':600}

# initialize grid search
gsearch = H2OGridSearch(H2ODeepLearningEstimator,
                        hyper_params=hyper_parameters,
                        search_criteria=search_criteria)

# execute training w/ grid search
gsearch.train(x=X,
              y=y,
              training_frame=train,
              validation_frame=validation)

# view detailed results at http://localhost:54321/flow/index.html
```

```python
# show grid search results
gsearch.show()

# select best model
nn_model2 = gsearch.get_grid()[0]

# print model information
nn_model2
```

```python
# measure nn AUC
print(nn_model2.auc(train=True))
print(nn_model2.auc(valid=True))
print(nn_model2.model_performance(test_data=test).auc())
```
